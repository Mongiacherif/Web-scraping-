## What is Web Scraping

```
#The librairies that we need to extract information from the link 
import requests 
import urllib
from bs4 import BeautifulSoup
import pandas as pd 
import numpy as np
```

## Part 1


The first part of my job, I will extract the information of such job and i will extract also the link description of each job for the same page.

First of all i represent the site that i choose to extract it ==> CRAIGSLIST.

Craigslist is an American classified advertisements website with sections devoted to jobs, housing, for sale, items wanted, services, community service, gigs, résumés, and discussion forums.  


```
#the link where we will extract information 
# For the scraping we choose craigList site to extract information
html ="https://boston.craigslist.org/search/edu?"
```

```
#We help us to see the source code html in format text 
html = urllib.request.urlopen(html).read()
print(html)
```

```
#convert text to format html
soup=BeautifulSoup(html,'html.parser')
print(soup)
```

## Extract link from the Craigslist

```
#Extract link 
link=soup.find_all('a')
for tag in link :
  print(tag.get('href'))
```

##Extract the titles of jobs available


```
#Extract the titles of jobs available in Boston from Craigslist  
  #result-title 
Title= soup.find_all(class_ ='result-title')
for title  in Title :
      print(title.text)
```

##Extract the adresses of jobs available 

```
#Extract the adresses of jobs available in Boston from Craigslist  
  #result-hood
Adresses= soup.find_all(class_ = 'result-hood')

for adress  in Adresses:
    print(adress.text)
```

##Extract Info of jobs

```
#Extract Info of jobs 
Jobs= soup.find_all("p",{'class' :'result-info'})

    
```

##Extract job details 


```
#Extract job details : we put it together 
 for job  in Jobs:
    Title= job.find("a",{'class' :'result-title'}).text
    Adresses_tag= job.find("span",{'class' :'result-hood'})
    Adresses= Adresses_tag.text if Adresses_tag else "N/A"
    Date= job.find("time",{'class' :'result-date'}).text
    Link= job.find("a",{'class' :'result-title'}).get('href')
    print('job title:',Title,'\nDate:',Date ,'\n Adress:',Adresses,'\n Link:',Link,'\n--------- ')
```

#extract information from the description link


```
#The next step 
#We have to extract information from the decription link 
#Here we just look for the same page 
jobs={}
jobs_nbre=0

for job  in Jobs:
    Title= job.find("a",{'class' :'result-title'}).text
    Adresses_tag= job.find("span",{'class' :'result-hood'})
    Adresses= Adresses_tag.text if Adresses_tag else "N/A"
    Date= job.find("time",{'class' :'result-date'}).text
    Link= job.find("a",{'class' :'result-title'}).get('href')
    job_link=urllib.request.urlopen(Link)
    job_soup=BeautifulSoup(job_link,'html.parser')
    job_description=job_soup.find('section',{'id': 'postingbody'}).text
    job_attributes_tag=job_soup.find('p',{'class':"attrgroup"})
    job_attributes= job_attributes_tag.text if job_attributes_tag else "N/A"
    jobs_nbre+=1
    jobs[jobs_nbre]=[Title,Adresses ,Date,Link,job_description,job_attributes]
   # dict={'Title':[Title],'Adresses':[Adresses],'Date':[Date],'Link':[Link],'job_description':[job_description],'job_attributes':[job_attributes]}
    print('job title:',Title,'\nDate:',Date ,'\n Adress:',Adresses,'\n Link:',Link,'\njob_description:',job_description,'\njob_attributes:',job_attributes,'\n--------- ')
```

After all the job we save our data into DataFrame 

```
#Creat a DataFrame contain our data 
job_df=pd.DataFrame.from_dict(jobs,orient='index',columns=['Title','Adresses ','Date','Link','job_description','job_attributes'])
```

# Part 2

The part 2 of my job i will extract from the website Craigslist from the field of job speciality general labo : https://boston.craigslist.org/search/lab to extract more pages 


The difference between the first part and this one here we wil extract more information from more pages and the same idea we will extract the link of description of our data. 

```
url=urllib.request.Request("https://boston.craigslist.org/search/lab",headers = {'User-Agent': 'Mozilla/5.0'})
#How we extract the jobs from the next page 
dic_job={}
job_nbre=0
while True:
  html= urllib.request.urlopen(url).read()
  soup=BeautifulSoup(html,'html.parser')
  #print(soup_1)
  Jobs= soup.find_all("p",{'class' :'result-info'})

  for job  in Jobs:
          Title= job.find("a",{'class' :'result-title'}).text
          Adresses_tag= job.find("span",{'class' :'result-hood'})
          Adresses= Adresses_tag.text if Adresses_tag else "N/A"
          Date= job.find("time",{'class' :'result-date'}).text
          Link= job.find("a",{'class' :'result-title'}).get('href')
          job_link=urllib.request.urlopen(Link)
          job_soup=BeautifulSoup(job_link,'html.parser')
          job_description=job_soup.find('section',{'id': 'postingbody'}).text
          job_attributes_tag=job_soup.find('p',{'class':"attrgroup"})
          job_attributes= job_attributes_tag.text if job_attributes_tag else "N/A"
        
          print('job title:',Title,'\nDate:',Date ,'\n Adress:',Adresses,'\n Link:',Link,'\njob_description:',job_description,'\njob_attributes:',job_attributes,'\n--------- ')
          job_nbre +=1
          dic_job[job_nbre]=[Title,Date ,Adresses,Link,job_description,job_attributes]
  url_tag=soup.find('a',{'title':'next page'})  
  if url_tag.get('href'):
      url="https://boston.craigslist.org/search/lab" + url_tag.get('href')
      print(url)
  else:
      break  
```

#Part 3


In this part the idea is to extract more and more information so 
i have reached more than 3000 informations from the same website Craiglists but this time i used the field of housing not a job like the others parts using different method.

We will extract the post of hosing in new hampshire from this website
 "https://nh.craigslist.org/search/hhh?"

```
#import get to call a get request on the site
from requests import get

#get the first page of the east bay housing prices
html = get('https://nh.craigslist.org/search/hhh?') 
html_soup = BeautifulSoup(html.text, 'html.parser')
result_row= html_soup.find_all('li', class_= 'result-row')
#we got 120 (elements/page)
print('the number of posts for 1 pages :',len(result_row)) 
results_num = html_soup.find('div', class_= 'search-legend')
#The total number of post 
total_number = results_num.find('span', class_='totalcount').text
total_number =int(total_number)
print('the number of posts for all pages :',total_number) 
#We have each page contain 120 post 
pages = np.arange(0, total_number+1, 120)
print('the page that we will extract all along this part :',pages)
```

```
import numpy as np

Date = []
neighborhood= []
Titles = []
Url= []
prices = []
bedroom=[]
area=[]


for page in pages:
    for post in result_row:

        if post.find('span', class_ = 'result-hood') is not None:

            #result-date
            
            result_date = post.find('time', class_= 'result-date')['datetime']
            Date.append(result_date)

            #neighborhoods
            
            result_hood = post.find('span', class_= 'result-hood').text
            neighborhood.append(result_hood)

            #title text
            
            result_title = post.find('a', class_='result-title hdrlnk')
            post_title = result_title.text
            Titles.append(post_title)

            #post link
            
            url = result_title['href']
            Url.append(url)

            #prices
            
            result_price = post.a.text.strip().replace("$", "")
            prices.append(result_price)

#The idea here not all the post contain the number of bedroom and the area 
#so if the post contain the class_ = 'housing' so we can enter in the post to extract either the surface or the number of bedrooms
# i put an example to better understand the following of my job 
# ['1br', '-', '600ft2', '-']
# 1br the number of bedroom , 600ft2 and this one for the area 

            
            if post.find('span', class_ = 'housing') is not None:
               
               #if the lengh of list is greater than 2 so we have in the post the number of bed and the area so we should to extract both 
               if len(post.find('span', class_ = 'housing').text.split()) > 2:
     
                    
                    num_bedroom = post.find('span', class_ = 'housing').text.replace("br", "").split()[0]
                    
                   
                    bedroom.append(num_bedroom)

                    
                    #for the area 
                    area_ = int(post.find('span', class_ = 'housing').text.split()[2][:-3])
                    
                    area.append(area_)

              #We have only the number of bedroom 
               elif len(post.find('span', class_ = 'housing').text.split()) == 2:
                    
                    #we extract the first case in the list 
                    num_bedroom = post.find('span', class_ = 'housing').text.replace("br", "").split()[0]
                    
                    
                    bedroom.append(num_bedroom)

                    #nan value 
                    area_ = np.nan
                    area.append(area_) 

               elif 'ft2' in post.find('span', class_ = 'housing').text.split()[0]:
                    
                    #make bedroom nan
                    num_bedroom = np.nan
                    bedroom.append(num_bedroom)
                    
                    #ok for area
                    area_ = int(post.find('span', class_ = 'housing').text.split()[0][:-3])
                    area.append(area_)

            else:
                num_bedroom = np.nan
                bedroom.append(num_bedroom)
                
                area_ = np.nan
                area.append(area_)    


                    
                  
```

```

import pandas as pd

df3 = pd.DataFrame({'Date of post': Date,'neighborhood': neighborhood,'post title': Titles,'URL': Url,'Prices' : prices,'bedroom':bedroom,'area':area})
```
